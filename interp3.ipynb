## Polinômios interpoladores como aproximantes de funções.

Uma das funções dos polinômios interpoladores é aproximar funções. Seja $f: [a,b] \to \mathbb{R}$ uma função de classe $\mathcal{C}^l$, e $\{x_0,\dots x_k\}$ um conjunto de pontos de $[a,b]$. Contruimos a tabela $T_f=\{(x_0,f(x_0)), \dots ,(x_k,f(x_k))\}$ e avaliaremos como é a aproximação de $f$ por $p_{T_f}$ (polinômio interpolador da tabela) no intervalo $[x_0,x_k]$. Ou seja: como é a função $f(x)-p_{T_f}(x)$ para $x\in[x_0,x_k]$?

### fórmula do erro
Consideremos $x_m \in [x_0,x_k]$ e $x_m \not\in \{x_0,\dots,x_k\}$. Note que $f(x_i)-p_{T_f}(x_i)=0$!
Defina as funções:
$$ \begin{gather} H(t) = (t-x_0)\cdots(t-x_k) \\
R(x) = f(x)-p_{T_f}(x)  \\
U(s) = R(s)H(x_m) - R(x_m)H(s) \end{gather} $$
Se $l\geq k+1$ temos que $U^{(k+1)}(s)$ tem um zero em $[x_0,x_1]$ e portanto existe $s_m$ talque $U^{(k+1)}(s_m)=0$ e
isto implica que:
$$ \begin{gather}
0 = R^{(k+1)}(s_m)H(x_m) - (k+1)!R(x_m) \\
R(x_m) = \frac{f^{(k+1)}(s_m)}{(k+1)!}(x_m-x_0)\cdots (x_m-x_k) \end{gather} $$

### Tabela de diferenças simples

Suponha que na tabela $T_f$ os pontos esteja igualmente espaçados, isto é, temos que
$x_i = x_0 + ih$, para algum $h>0$. Neste caso o cálculo das diferenças divididas não vai depender dos $x_i$.
Definimos recursivamente 
$$ \begin{gather} \Delta^l f(x_i) = \Delta^{l-1}f(x_i) - \Delta^{l-1}f(x_{i-1})\\
\Delta^1 f(x_i) = f(x_i) - f(x_{i-1})\end{gather}$$
A fórmula de diferenças divididas então fica
$$ f[x_0,\dots,x_l] = \frac{\Delta^lf(x_l)}{l!h^l} $$
### Exemplos


```python
%matplotlib inline
import numpy as np
x = np.linspace(-3.0, 3.0, 5)
y = np.sin(x)
p = np.polyfit(x,y,4)
```

```python
from matplotlib.pyplot import plot,grid
p4 = lambda x: np.polyval(p,x)
plot(x,y,"ro")
t=np.linspace(-3,3,300)
plot(t,np.sin(t),t,p4(t))
grid()
```

```python
# Variação dos numeros de pontos na tabela
x1 = np.linspace(-5,5,11)
f= lambda x : 1/(1+x**2)
y1 = f(x1)
p10 = np.polyfit(x1,y1,10)
u = lambda t: np.polyval(p10,t)
t=np.linspace(-5,5,300)
plot(t,f(t),t,u(t))
```

## Polinômios de Tchebitchev
Considere a função peso no intervalo $(-1,1)$  definida por $w(x) = \frac{1}{\sqrt{1-x^2}}$, e a família de polinômios ortogonais em relação ao produto interno:
$$ \langle f,g\rangle = \int_{-1}^1w(x)f(x)g(x)dx $$
Fazendo a mudança de variável $x=\cos(\theta)$ com $\theta \in [0,\pi]$ temos $\langle f,g\rangle = \int_0^{\pi}f(\cos(\theta))g(\cos(\theta))d\theta$, assim os dois primeiros polinômios ortogonais são

* $T_0(x) = 1$
* $T_1(x) = x$
* $T_2(x) = 2x^2 -1$

### Teorema 
$T_{n+1}(x) = 2xT_n(x)- T_{n-1}(x)$

Note que $T_n(\cos(\theta)) = \cos(n\theta)$.

```python
# Polinômios de Chebichev em numpy.
import numpy.polynomial.chebyshev as cheb
T0= lambda x: cheb.chebval(x,[1])
T1= lambda x: cheb.chebval(x,[0,1])
T2 = lambda x: cheb.chebval(x,[0,0,1])
T4 = lambda x : cheb.chebval(x,[0,0,0,0,1])
raizes = cheb.chebroots([0,0,0,0,1])
```

```python
t = np.linspace(-1,1,300)
plot(t,T0(t),t,T1(t),t,T2(t),t,T4(t))
```
